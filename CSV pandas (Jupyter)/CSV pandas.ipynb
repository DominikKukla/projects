Code not so much optimized, done to help me memorize basic commands used to manipulate CSV files with the help of the Pandas module

Importing necessary libraries and CSV file

```python
import pandas as pd
import winsound as ws
import time as tm
people = pd.read_csv('people.csv')
```

Function receives table and row index and checks whether a row exists under that index

```python
def row_index_exists(table, row_index):
    if int(row_index) not in table.index:
        return False
    return True
```

Function receives table and row number and checks whether a row exists under that number

```python
def row_number_exists(table, row_number):
    if int(row_number) > len(table.index) - 1 or int(row_number) < len(table.index) * -1:
        return False
    return True
```

Function receives table and column name and checks whether a column exists under that name

```python
def column_name_exists(table, column_name):
    if column_name not in table.columns:
        return False
    return True
```

Function receives table and column number and checks whether a column exists under that number

```python
def column_number_exists(table, column_number):
    if int(column_number) > len(table.columns) - 1 or int(column_number) < len(table.columns) * -1:
        return False
    return True
```

Function receives table and displays first n rows of it given by the user

```python
def display_table(table):
    n = input("Input integer value n:\n>>> ")
    if n.isdigit():
        display(table.head(int(n)))
        tm.sleep(0.1)
    else:
        print("n must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
```

Function receives table and displays data from .describe() method of column given by the user

```python
def display_describe(table):
    column_name = input("Input column name:\n>>> ")
    if column_name_exists(table, column_name):
        data = table[column_name].describe()
        print("Values count:", data[0])
        print("Unique values count:", data[1])
        print("The most frequent value:", data[2])
        print("Count of it's occurrences", data[3])
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
```

Function receives table and sorts it ascending or descending chosen by the user

```python
def sort(table):
    column_name = input("Input column name:\n>>> ")
    if column_name_exists(table, column_name):
        sorting_type = input("Select sorting type: a-z | z-a:\n>>> ")
        if sorting_type == "a-z":
            table = table.sort_values(by=column_name, ascending=True)
            display(table)
            tm.sleep(0.1)
            return table
        elif sorting_type == "z-a":
            table = table.sort_values(by=column_name, ascending=False)
            display(table)
            tm.sleep(0.1)
            return table
        else:
            print("Such sorting type is non-existent!")
            ws.PlaySound("beep.wav", ws.SND_FILENAME)
            return table
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
```

Function receives table and displays data using row indexes and column names given by the user

```python
def specific_indexes(table):
    row_indexes = []
    print("Input row indexes | to move on input 'stop'")
    while True:
        row_index = input(">>> ")
        if row_index != "stop":
            if row_index.replace('-', '').isdigit():
                if row_index_exists(table, row_index):
                    row_indexes.append(int(row_index))
                else:
                    print("Such row is non-existent!")
                    ws.PlaySound("beep.wav", ws.SND_FILENAME)
            else:
                print("Row index must be an integer!")
                ws.PlaySound("beep.wav", ws.SND_FILENAME)
        else:
            break
    column_names = []
    print("Input column names | to move on input 'stop'")
    while True:
        column_name = input(">>> ")
        if column_name != "stop":
            if column_name_exists(table, column_name):
                column_names.append(column_name)
            else:
                print("Such column is non-existent!")
                ws.PlaySound("beep.wav", ws.SND_FILENAME)
        else:
            break
    display(table.loc[row_indexes, column_names])
    tm.sleep(0.1)
```

Function receives table and displays data using row numbers and column numbers given by the user

```python
def specific_row_numbers(table):
    row_numbers = []
    print("Input row numbers | to move on input 'stop'")
    while True:
        row_number = input(">>> ")
        if row_number != "stop":
            if row_number.replace('-', '').isdigit():
                if row_number_exists(table, row_number):
                    row_numbers.append(int(row_number))
                else:
                    print("Such row is non-existent!")
                    ws.PlaySound("beep.wav", ws.SND_FILENAME)
            else:
                print("Row number must be an integer!")
                ws.PlaySound("beep.wav", ws.SND_FILENAME)
        else:
            break
    column_numbers = []
    print("Input column numbers | to move on input 'stop'")
    while True:
        column_number = input(">>> ")
        if column_number != 'stop':
            if column_number.replace('-', '').isdigit():
                if column_number_exists(table, column_number):
                    column_numbers.append(int(column_number))
                else:
                    print("Such columns is non-existent!")
                    ws.PlaySound("beep.wav", ws.SND_FILENAME)
            else:
                print("Column number must be an integer!")
                ws.PlaySound("beep.wav", ws.SND_FILENAME)
        else:
            break
    display(table.iloc[row_numbers, column_numbers])
    tm.sleep(0.1)
```

Function receives table and displays data using range of row indexes and range of column names given by the user

```python
def range_of_indexes(table):
    index_start = input("Input starting index:\n>>> ")
    if index_start.replace('-', '').isdigit():
        if row_index_exists(table, index_start):
            pass
        else:
            print("Such row is non-existent!")
            ws.PlaySound("beep.wav", ws.SND_FILENAME)
            return
    else:
        print("Row index must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return
    index_end = input("Input ending index:\n>>> ")
    if index_end.replace('-', '').isdigit():
        if row_index_exists(table, index_end):
            pass
        else:
            print("Such row is non-existent!")
            ws.PlaySound("beep.wav", ws.SND_FILENAME)
            return
    else:
        print("Row index must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return
    column_name_start = input("Input starting column name:\n>>> ")
    if column_name_exists(table, column_name_start):
        pass
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return
    column_name_end = input("Input ending column name:\n>>> ")
    if column_name_exists(table, column_name_end):
        pass
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return
    display(table.loc[int(index_start):int(index_end), column_name_start:column_name_end])
    tm.sleep(0.1)
```

Function receives table and displays data using range of row numbers and range of column numbers given by the user

```python
def range_of_number_rows(table):
    row_number_start = input("Input starting row number:\n>>> ")
    if row_number_start.replace('-', '').isdigit():
        if row_number_exists(table, row_number_start):
            pass
        else:
            print("Such row is non-existent!")
            ws.PlaySound("beep.wav", ws.SND_FILENAME)
            return
    else:
        print("Row number must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return
    row_number_end = input("Input ending row number:\n>>> ")
    if row_number_end.replace('-', '').isdigit():
        if row_number_exists(table, row_number_end):
            pass
        else:
            print("Such row is non-existent!")
            ws.PlaySound("beep.wav", ws.SND_FILENAME)
            return
    else:
        print("Row number must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return
    column_number_start = input("Input starting column number:\n>>> ")
    if column_number_start.replace('-', '').isdigit():
        if column_number_exists(table, column_number_start):
            pass
        else:
            print("Such column is non-existent!")
            ws.PlaySound("beep.wav", ws.SND_FILENAME)
            return
    else:
        print("Column number must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return
    column_number_end = input("Input ending column number:\n>>> ")
    if column_number_end.replace('-', '').isdigit():
        if column_number_exists(table, column_number_end):
            pass
        else:
            print("Such column is non-existent!")
            ws.PlaySound("beep.wav", ws.SND_FILENAME)
            return
    else:
        print("Column number must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return
    display(table.iloc[int(row_number_start):int(row_number_end), int(column_number_start):int(column_number_end)])
    tm.sleep(0.1)
```

Function receives table and displays possible methods of displaying data from table and reads the user choice

```python
def display_method(table):
    method = input("""Pick what method of displaying data you want to use:
1 - Specific indexes and specific column names
2 - Specific row numbers and specific column numbers
3 - Range of indexes and range of column names (residual value included)
4 - Range of row number and range of column numbers (residual value not included)
>>> """)
    if method == "1":
        specific_indexes(table)
    elif method == "2":
        specific_row_numbers(table)
    elif method == "3":
        range_of_indexes(table)
    elif method == "4":
        range_of_number_rows(table)
    else:
        print("Such method is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
```

Function receives table and displays table values found by fraze given by the user

```python
def search(table):
    column = input("Input column name in which you want to search for phrase:\n>>> ")
    if column_name_exists(table, column):
        phrase = input("Input searched phrase:\n>>> ")
        display(table[table[column].str.find(phrase) != -1])
        tm.sleep(0.1)
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
```

Function receives table, changes column name to given by the user and displays table

```python
def column_name_change(table):
    column = input("Input column name:\n>>> ")
    if column_name_exists(table, column):
        changed_name = input("Input changed column name:\n>>> ")
        table = table.rename(mapper={column: changed_name}, axis=1)
        display(table)
        tm.sleep(0.1)
        return table
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
```

Function receives table, deletes column given by the user and displays table

```python
def delete_column(table):
    column = input("Input column name:\n>>> ")
    if column_name_exists(table, column):
        table = table.drop([column], axis=1)
        display(table)
        tm.sleep(0.1)
        return table
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
```

Function receives table, creates new column with name and values given by the user and displays table

```python
def create_column(table):
    column = input("Input column name:\n>>> ")
    value = input("Input default value:\n>>> ")
    if_int = False
    if_float = False
    message = "Pick value type:\n1 - String\n"
    if value.replace('-', '').isdigit():
        if_int = True
        message += "2 - Integer\n"
    if value.replace('-', '').replace('.', '', 1).isdigit():
        if_float = True
        message += "3 - Floating point\n"
    type_ = input(message+">>> ")
    if type_ == "1":
        table[column] = str(value)
        display(table)
        tm.sleep(0.1)
        return table
    elif type_ == "2" and if_int:
        table[column] = int(value)
        display(table)
        tm.sleep(0.1)
        return table
    elif type_ == "3" and if_float:
        table[column] = float(value)
        display(table)
        tm.sleep(0.1)
        return table
    else:
        print("Such value type is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
```

Function receives table, splits values in column by separator given by the user and assigns them into new columns

```python
def split_column(table):
    column = input("Input column name:\n>>> ")
    if column_name_exists(table, column):
        pass
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
    pattern = input("Input the separator:\n>>> ")
    data = table[column].str.split(pat=pattern, expand=True)
    print("Number of split values:", len(data.columns))
    for value in range(0, len(data.columns)):
        pick = input(f"""Input the destination column for the values: {value + 1}?
Input 'skip' to omit a value
>>> """)
        if pick == "skip":
            continue
        table[pick] = data[value]
    display(table)
    tm.sleep(0.1)
    return table
```

Function receives table, concatenates values from two columns into one by the separator given by the user and assigns it into new column

```python
def concat_column(table):
    column1 = input("Input first column name:\n>>> ")
    if column_name_exists(table, column1):
        pass
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
    column2 = input("Input second column name:\n>>> ")
    if column_name_exists(table, column2):
        pass
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
    separator = input("Input the separator:\n>>> ")
    new = input("Input the destination column for the value:\n>>> ")
    table[new] = table[column1] + separator + table[column2]
    display(table)
    tm.sleep(0.1)
    return table
```

Function receives table, checks whether a column has in itself integers or floats and rounds them to n decimal digits given by the user

```python
def round_column(table):
    column = input("Input column name:\n>>> ")
    if column_name_exists(table, column):
        pass
    else:
        print("Such column is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
    digit = input("Input the number of decimal digits to round:\n>>> ")
    if digit.isdigit():
        pass
    else:
        print("The number of decimal digits must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
    if table[column].dtype == "int64":
        table[column] = table[column].astype(float)
    elif table[column].dtype == "float64":
        pass
    else:
        print("Column must consist of digit values only!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
    table[column] = table[column].round(decimals=int(digit))
    display(table)
    tm.sleep(0.1)
    return table
```

Function receives table and changes value of cell indicated by row index and column name given by the user

```python
def value_change(table):
    column = input("Input column name:\n>>> ")
    row_index = input("Input row index:\n>>> ")
    value = input("Input new column value:\n>>> ")
    if row_index.replace('-', '').isdigit():
        pass
    else:
        print("Index must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
    table.at[int(row_index), column] = value
    display(table)
    tm.sleep(0.1)
    return table
```

Function receives table and deletes row by index given by the user

```python
def delete_row(table):
    row_index = input("Input row index:\n>>> ")
    if row_index.replace('-', '').isdigit():
        if row_index_exists(table, row_index):
            pass
        else:
            print("Such row is non-existent!")
            ws.PlaySound("beep.wav", ws.SND_FILENAME)
            return table
    else:
        print("Row index must be an integer!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
        return table
    table = table.drop(index=int(row_index), axis=0)
    display(table)
    tm.sleep(0.1)
    return table
```

Function receives table and resets it

```python
def reset_table(table):
    safe_check = input("Are you sure you want to reset the table? YES | NO\n>>> ")
    if safe_check == "YES":
        print("Reset done.")
        table = pd.read_csv('people.csv')
        display(table)
        tm.sleep(0.1)
        return table
    else:
        return table
```

Function displays possible actions related to the table | Displays the table | 
Main program loop

```python
display(people)
tm.sleep(0.1)
while True:
    print("----------------------------------------------------")
    choice = input("""
Please pick desirable action:
1.1 - Display n rows
1.2 - Display column summary
1.3 - Display sorted by column
1.4 - Display data from column (4 methods)
1.5 - Display by searched phrase
2.1 - Change column name and display
2.2 - Delete column and display
2.3 - Add or reset column and display
2.4 - Split column and display
2.5 - Concatenate two columns and display
2.6 - Round column to n decimal digits and display
3.1 - Change value of the cell and display
3.2 - Delete row and display
del - Reset table
end - Quit program
>>> """)
    if choice == "1.1":
        display_table(people)
    elif choice == "1.2":
        display_describe(people)
    elif choice == "1.3":
        people = sort(people)
    elif choice == "1.4":
        display_method(people)
    elif choice == "1.5":
        search(people)
    elif choice == "2.1":
        people = column_name_change(people)
    elif choice == "2.2":
        people = delete_column(people)
    elif choice == "2.3":
        people = create_column(people)
    elif choice == "2.4":
        people = split_column(people)
    elif choice == "2.5":
        people = concat_column(people)
    elif choice == "2.6":
        people = round_column(people)
    elif choice == "3.1":
        people = value_change(people)
    elif choice == "3.2":
        people = delete_row(people)
    elif choice == "del":
        people = reset_table(people)
    elif choice == "end":
        print("Quited.")
        break
    else:
        print("Such action is non-existent!")
        ws.PlaySound("beep.wav", ws.SND_FILENAME)
    print()
```
